```python
fig, ax = plt.subplots(ncols=8, nrows=4, figsize=(30,15))

for i in range(4):
    for j in range(8):
        if (i*8+j > 13 and i*8+j < 22):
            _, _, hist_tt = ax[i,j].hist([x for x in df_tt[features[i*8+j]].to_numpy() if x != 0.], bins=50, density=True, histtype='step')
            _, _, hist_tl = ax[i,j].hist([x for x in df_tl[features[i*8+j]].to_numpy() if x != 0.], bins=50, density=True, histtype='step')
            _, _, hist_ll = ax[i,j].hist([x for x in df_ll[features[i*8+j]].to_numpy() if x != 0.], bins=50, density=True, histtype='step')
        else:
            ax[i,j].hist(df_tt[features[i*8+j]].to_numpy(), bins=50, density=True, histtype='step')
            ax[i,j].hist(df_tl[features[i*8+j]].to_numpy(), bins=50, density=True, histtype='step')
            ax[i,j].hist(df_ll[features[i*8+j]].to_numpy(), bins=50, density=True, histtype='step')
        ax[i,j].set_title(features[i*8+j])
        #if log_mask[i*8+j]: ax[i,j].set_xscale('log')

fig.suptitle('Features')
fig.legend([hist_tt[0], hist_tl[0], hist_ll[0]], ['TT', 'TL', 'LL'], loc="upper center") 
plt.legend()
```
