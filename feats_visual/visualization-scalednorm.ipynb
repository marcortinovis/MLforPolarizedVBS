```python
binss_tt = []
binss_tl = []
binss_ll = []

edgess_tt = []
edgess_tl = []
edgess_ll = []
```

```python
fig, ax = plt.subplots(ncols=8, nrows=4, figsize=(20,10))

for i in range(4):
    for j in range(8):
        if (i*8+j > 13 and i*8+j < 22):
            bins_tt, edges_tt, hist_tt = ax[i,j].hist([x for x in df_tt[features[i*8+j]].to_numpy() if x != 0.], bins=50, density=True, histtype='step')
            bins_tl, edges_tl, hist_tl = ax[i,j].hist([x for x in df_tl[features[i*8+j]].to_numpy() if x != 0.], bins=50, density=True, histtype='step')
            bins_ll, edges_ll, hist_ll = ax[i,j].hist([x for x in df_ll[features[i*8+j]].to_numpy() if x != 0.], bins=50, density=True, histtype='step')
        else:
            bins_tt, edges_tt, hist_tt = ax[i,j].hist(df_tt[features[i*8+j]].to_numpy(), bins=50, density=True, histtype='step')
            bins_tl, edges_tl, hist_tl = ax[i,j].hist(df_tl[features[i*8+j]].to_numpy(), bins=50, density=True, histtype='step')
            bins_ll, edges_ll, hist_ll = ax[i,j].hist(df_ll[features[i*8+j]].to_numpy(), bins=50, density=True, histtype='step')
        ax[i,j].set_title(features[i*8+j])
        #if log_mask[i*8+j]: ax[i,j].set_xscale('log')
        binss_tt.append(bins_tt)
        binss_tl.append(bins_tl)
        binss_ll.append(bins_ll)
        edgess_tt.append([0.5*(e1+e2) for e1, e2 in zip(edges_tt[:-1], edges_tt[1:])])
        edgess_tl.append([0.5*(e1+e2) for e1, e2 in zip(edges_tl[:-1], edges_tl[1:])])
        edgess_ll.append([0.5*(e1+e2) for e1, e2 in zip(edges_ll[:-1], edges_ll[1:])])

fig.suptitle('Features extracted from ROOT data (SSWW_polarized_CMrf.root): three classes TT, TL, LL, normalized histograms')
fig.legend([hist_tt[0], hist_tl[0], hist_ll[0]], ['TT', 'TL', 'LL'], loc="upper right", bbox_to_anchor=(0.9,1.))
plt.legend()
plt.tight_layout()
plt.savefig('scaled_features.png')
```

```python
pulls_tl = []
pulls_tt = []
pulls_tx = []
for bins_ll in binss_ll:
    pulls_tl.append([(a-b)/b for a, b in zip(bins_ll, bins_tl)])
    pulls_tt.append([(a-b)/b for a, b in zip(bins_ll, bins_tt)])
    pulls_tx.append([(a-b)/b for a, b in zip(bins_ll, bins_tt+bins_tl)])
```

```python
fig, ax = plt.subplots(ncols=4, nrows=8, figsize=(20,20))

for i in range(8):
    for j in range(4):
        ax[i,j].scatter(edgess_tl[i*4+j], pulls_tl[i*4+j], s=5)
        ax[i,j].scatter(edgess_tt[i*4+j], pulls_tt[i*4+j], s=5)
        ax[i,j].set_title(features[i*4+j])
        ax[i,j].hlines(0., min(edgess_tl[i*4+j]), max(edgess_tl[i*4+j]), color='k', linestyle='dashed', alpha=0.75)
        ax[i,j].set_ylim(-5, 5)
        #if log_mask[i*8+j]: ax[i,j].set_xscale('log')
fig.legend([hist_tl[0], hist_tt[0]], ['TT', 'TL'], loc="upper right", bbox_to_anchor=(0.9,1.))
plt.legend()
plt.tight_layout()
```
