```python
el_mass = 0.000510998950
mu_mass = 0.10565837
w_mass = 80.3692
nu_mass = 0.

def four_momentum(l_id, pt, eta, phi):
    # only for electrons or muons
    # from pt, eta, phi, m to en, px, py, pz
    px = pt * np.cos(phi)
    py = pt * np.sin(phi)
    pz = pt * np.sinh(eta)
    p = pt * np.cosh(eta)
    en = np.where(np.abs(l_id)==11, np.sqrt(el_mass**2 + p**2), np.sqrt(mu_mass**2 + p**2))
    return np.array([en, px, py, pz]).T

def back_four_momentum(fourp):
    # only for w
    # from en, px, py, pz to pt, eta, phi
    en, px, py, pz = fourp[:,0], fourp[:,1], fourp[:,2], fourp[:,3]
    p = np.sqrt(px**2 + py**2 + pz**2)
    theta = np.arccos(pz/p)
    eta = - np.log(np.tan(theta*0.5))
    pt = np.sqrt(px**2 + py**2)
    phi = np.arccos(px/pt) - np.pi*np.heaviside(-py, 0)
    en = np.sqrt(w_mass**2 + p**2)
    return np.array([pt, eta, phi]).T

def inv_mass(fourv):
    return np.sqrt(fourv[:,0]**2 - fourv[:,1]**2 - fourv[:,2]**2 - fourv[:,3]**2)
```

```python
def w_selection(leps, nus):
    
    # leptons 4-momenta
    lep1 = leps[:,::2]
    lep2 = leps[:,1::2]
    lep1_4p = four_momentum(lep1[:,0], lep1[:,1], lep1[:,2], lep1[:,3])
    lep2_4p = four_momentum(lep2[:,0], lep2[:,1], lep2[:,2], lep2[:,3])
    nu1 = nus[:,::2]
    nu2 = nus[:,1::2]
    nu1_4p = four_momentum(nu_mass, nu1[:,0], nu1[:,1], nu1[:,2])
    nu2_4p = four_momentum(nu_mass, nu2[:,0], nu2[:,1], nu2[:,2])
    
    # invariant masses of the possible couples
    sum11 = lep1_4p+nu1_4p
    sum22 = lep2_4p+nu2_4p
    sum12 = lep1_4p+nu2_4p
    sum21 = lep2_4p+nu1_4p
    invm_sum11 = inv_mass(sum11)
    invm_sum22 = inv_mass(sum22)
    invm_sum12 = inv_mass(sum12)
    invm_sum21 = inv_mass(sum21)
    invm_dev11 = np.array([abs(x-w_mass) for x in invm_sum11]) # deviation of the invariant mass from the expected w boson mass
    invm_dev22 = np.array([abs(x-w_mass) for x in invm_sum22])
    invm_dev12 = np.array([abs(x-w_mass) for x in invm_sum12])
    invm_dev21 = np.array([abs(x-w_mass) for x in invm_sum21])
    invm_dev1122 = np.add(invm_dev11, invm_dev22)
    invm_dev1221 = np.add(invm_dev12, invm_dev21)
    couple_selection = np.where(invm_dev1122 > invm_dev1221, 1, 0) # choose the couple as the one that is nearer to w_mass
    sums = np.array(list(zip(list(zip(sum11, sum22)), list(zip(sum12, sum21)))))
    invm_sums = np.array(list(zip(list(zip(invm_sum11, invm_sum22)), list(zip(invm_sum12, invm_sum21)))))
    selected_couples = []
    for i in range(len(invm_sums)):
        selected_couples.append(list(sums[i][couple_selection[i]]))
    selected_couples = np.array(selected_couples)
    res = back_four_momentum(selected_couples.reshape((len(selected_couples)*2,4))).reshape((len(selected_couples),2,3))
    ptW1 = res[:,0,0]
    etaW1 = res[:,0,1]
    phiW1 = res[:,0,2]
    ptW2 = res[:,1,0]
    etaW2 = res[:,1,1]
    phiW2 = res[:,1,2]
    return [ptW1, etaW1, phiW1], [ptW2, etaW2, phiW2], selected_couples
```

```python
W1_feats_ll, W2_feats_ll, selected_couples_ll = w_selection(leps_ll, nus_ll)
W1_feats_tl, W2_feats_tl, selected_couples_tl = w_selection(leps_tl, nus_tl)
W1_feats_tt, W2_feats_tt, selected_couples_tt = w_selection(leps_tt, nus_tt)
```

```python
m_el = 0.000510998950
m_mu = 0.10565837

def lepton_boosting(leps, W1, W2, selected_couples):
    # boosts the leptons into the W rf, both rotating the axes beforehand (such that the new z axis is in the w lab direction) and without that step
    # (we think the first is more correct)
    
    w_mass = 80.3692

    # Ws
    W_directions = selected_couples[:,:,1:]
    W1_dirs = W_directions[:,0]
    W2_dirs = W_directions[:,1]

    z_axis = ROOT.Math.XYZVector(0., 0., 1.)

    W1_dirs = [ROOT.Math.XYZVector(w[0], w[1], w[2]) for w in W1_dirs]
    W2_dirs = [ROOT.Math.XYZVector(w[0], w[1], w[2]) for w in W2_dirs]
    leps1_3p = np.array([ROOT.Math.XYZVector(l[1], l[2], l[3]) for l in leps[:,::2]])
    leps2_3p = np.array([ROOT.Math.XYZVector(l[1], l[2], l[3]) for l in leps[:,1::2]])
    # find axis and angles of rotation
    axes_1 = np.array([w.Cross(z_axis) for w in W1_dirs])
    axes_2 = np.array([w.Cross(z_axis) for w in W2_dirs])
    angles_1 = np.array([np.arccos(w.Dot(z_axis) / np.sqrt(w.Mag2() * z_axis.Mag2())) for w in W1_dirs])
    angles_2 = np.array([np.arccos(w.Dot(z_axis) / np.sqrt(w.Mag2() * z_axis.Mag2())) for w in W2_dirs])
    leps1_4p = np.array([ROOT.Math.PxPyPzMVector(l[1], l[2], l[3], m_el) if np.abs(l[0]) == 11. else ROOT.Math.PxPyPzMVector(l[1], l[2], l[3], m_mu) for l in leps[:,::2]])
    leps2_4p = np.array([ROOT.Math.PxPyPzMVector(l[1], l[2], l[3], m_el) if np.abs(l[0]) == 11. else ROOT.Math.PxPyPzMVector(l[1], l[2], l[3], m_mu) for l in leps[:,1::2]])
    rotations_1 = []
    rotations_2 = []
    print('Rotations found')
    for i in range(len(axes_1)):
        rotations_1.append(ROOT.Math.LorentzRotation(ROOT.Math.AxisAngle(axes_1[i], float(angles_1[i]))))
        rotations_2.append(ROOT.Math.LorentzRotation(ROOT.Math.AxisAngle(axes_2[i], float(angles_2[i]))))
    rotated_l4p_1 = np.array([r(l) for r, l in zip(rotations_1, leps1_4p)])
    rotated_l4p_2 = np.array([r(l) for r, l in zip(rotations_2, leps2_4p)])
    rotated_l4p_1 = np.array([[rl.X(), rl.Y(), rl.Z(), rl.mass()] for rl in rotated_l4p_1])
    rotated_l4p_2 = np.array([[rl.X(), rl.Y(), rl.Z(), rl.mass()] for rl in rotated_l4p_2])
    W1_4p = np.array([ROOT.Math.PxPyPzMVector(w.X(), w.Y(), w.Z(), w_mass) for w in W1_dirs])
    W2_4p = np.array([ROOT.Math.PxPyPzMVector(w.X(), w.Y(), w.Z(), w_mass) for w in W2_dirs])
    rotated_w1 = np.array([r(w) for r, w in zip(rotations_1, W1_4p)])
    rotated_w2 = np.array([r(w) for r, w in zip(rotations_2, W2_4p)])
    rotated_w1 = np.array([[rw.X(), rw.Y(), rw.Z(), rw.mass()] for rw in rotated_w1])
    rotated_w2 = np.array([[rw.X(), rw.Y(), rw.Z(), rw.mass()] for rw in rotated_w2])
    print('Rotations done')

    W1s = np.array([W1[0], W1[1], W1[2]]).T
    W2s = np.array([W2[0], W2[1], W2[2]]).T

    # search for betas
    betas_1 = []
    betas_2 = []
    for W1, W2 in zip(W1s, W2s):
        W1 = ROOT.Math.PtEtaPhiMVector(W1[0], W1[1], W1[2], w_mass)
        W2 = ROOT.Math.PtEtaPhiMVector(W2[0], W2[1], W2[2], w_mass)
        boost_1 = W1.BoostToCM()
        boost_2 = W2.BoostToCM()
        b_1_x, b_1_y, b_1_z = boost_1.x(), boost_1.y(), boost_1.z()
        b_2_x, b_2_y, b_2_z = boost_2.x(), boost_2.y(), boost_2.z()
        beta_1 = [b_1_x, b_1_y, b_1_z]
        beta_2 = [b_2_x, b_2_y, b_2_z]
        betas_1.append(beta_1)
        betas_2.append(beta_2)
    betas_1 = np.array(betas_1)
    betas_2 = np.array(betas_2)
    print('Boosts found')

    leps1_4p = np.array([[l[1], l[2], l[3], m_el] if np.abs(l[0]) == 11. else [l[1], l[2], l[3], m_mu] for l in leps[:,::2]])
    leps2_4p = np.array([[l[1], l[2], l[3], m_el] if np.abs(l[0]) == 11. else [l[1], l[2], l[3], m_mu] for l in leps[:,1::2]])
    leps1_4p = np.array([ROOT.Math.PtEtaPhiMVector(l[0], l[1], l[2], l[3]) for l in leps1_4p])
    leps2_4p = np.array([ROOT.Math.PtEtaPhiMVector(l[0], l[1], l[2], l[3]) for l in leps2_4p])
    rotated_l4p_1 = np.array([ROOT.Math.PxPyPzMVector(lep[0], lep[1], lep[2], lep[3]) for lep in rotated_l4p_1])
    rotated_l4p_2 = np.array([ROOT.Math.PxPyPzMVector(lep[0], lep[1], lep[2], lep[3]) for lep in rotated_l4p_2])
    betas_1 = [ROOT.TVector3(b[0], b[1], b[2]) for b in betas_1]
    betas_2 = [ROOT.TVector3(b[0], b[1], b[2]) for b in betas_2]
    boosts_1 = [ROOT.Math.Boost(-b) for b in betas_1]
    boosts_2 = [ROOT.Math.Boost(-b) for b in betas_2]

    # 'nr' stands for no-rotation
    boosted_nr_leps1_4p = np.array([ROOT.Math.PxPyPzMVector(boost(lep)) for boost, lep in zip(boosts_1, leps1_4p)])
    boosted_nr_leps2_4p = np.array([ROOT.Math.PxPyPzMVector(boost(lep)) for boost, lep in zip(boosts_2, leps2_4p)])
    boosted_nr_leps1_4p = np.array([[l.Px(), l.Py(), l.Pz(), l.mass()] for l in boosted_nr_leps1_4p])
    boosted_nr_leps2_4p = np.array([[l.Px(), l.Py(), l.Pz(), l.mass()] for l in boosted_nr_leps2_4p])
    
    boosted_leps1_4p = np.array([ROOT.Math.PxPyPzMVector(boost(lep)) for boost, lep in zip(boosts_1, rotated_l4p_1)])
    boosted_leps2_4p = np.array([ROOT.Math.PxPyPzMVector(boost(lep)) for boost, lep in zip(boosts_2, rotated_l4p_1)])
    boosted_leps1_4p = np.array([[l.Px(), l.Py(), l.Pz(), l.mass()] for l in boosted_leps1_4p])
    boosted_leps2_4p = np.array([[l.Px(), l.Py(), l.Pz(), l.mass()] for l in boosted_leps2_4p])
    print('Boosts done')

    costheta_1_nr = boosted_nr_leps1_4p[:,2]/np.sqrt(boosted_nr_leps1_4p[:,0]**2+boosted_nr_leps1_4p[:,1]**2+boosted_nr_leps1_4p[:,2]**2)
    costheta_2_nr = boosted_nr_leps2_4p[:,2]/np.sqrt(boosted_nr_leps2_4p[:,0]**2+boosted_nr_leps2_4p[:,1]**2+boosted_nr_leps2_4p[:,2]**2)
    costheta_1 = boosted_leps1_4p[:,2]/np.sqrt(boosted_leps1_4p[:,0]**2+boosted_leps1_4p[:,1]**2+boosted_leps1_4p[:,2]**2)
    costheta_2 = boosted_leps2_4p[:,2]/np.sqrt(boosted_leps2_4p[:,0]**2+boosted_leps2_4p[:,1]**2+boosted_leps2_4p[:,2]**2)

    return boosted_nr_leps1_4p, boosted_nr_leps2_4p, boosted_leps1_4p, boosted_leps2_4p, costheta_1_nr, costheta_2_nr, costheta_1, costheta_2
```

```python
boosted_nr_leps1_4p_ll, boosted_nr_leps2_4p_ll, boosted_leps1_4p_ll, boosted_leps2_4p_ll, costheta_1_nr_ll, costheta_2_nr_ll, costheta_1_ll, costheta_2_ll = lepton_boosting(leps_ll, W1_feats_ll, W2_feats_ll, selected_couples_ll)
print('LL done')

boosted_nr_leps1_4p_tl, boosted_nr_leps2_4p_tl, boosted_leps1_4p_tl, boosted_leps2_4p_tl, costheta_1_nr_tl, costheta_2_nr_tl, costheta_1_tl, costheta_2_tl = lepton_boosting(leps_tl, W1_feats_tl, W2_feats_tl, selected_couples_tl)
print('TL done')

boosted_nr_leps1_4p_tt, boosted_nr_leps2_4p_tt, boosted_leps1_4p_tt, boosted_leps2_4p_tt, costheta_1_nr_tt, costheta_2_nr_tt, costheta_1_tt, costheta_2_tt = lepton_boosting(leps_tt, W1_feats_tt, W2_feats_tt, selected_couples_tt)
print('TT done')
```

```python
# Scaling

def logit(x, clip=1e-6):
    x = np.clip(x, a_min=clip, a_max=1-clip)
    return np.log(x/(1-x))

scaled_costheta_1_ll = logit((costheta_1_ll + 1.)/(2.), clip=1e-6)
scaled_costheta_2_ll = logit((costheta_2_ll + 1.)/(2.), clip=1e-6)
scaled_costheta_1_tl = logit((costheta_1_tl + 1.)/(2.), clip=1e-6)
scaled_costheta_2_tl = logit((costheta_2_tl + 1.)/(2.), clip=1e-6)
scaled_costheta_1_tt = logit((costheta_1_tt + 1.)/(2.), clip=1e-6)
scaled_costheta_2_tt = logit((costheta_2_tt + 1.)/(2.), clip=1e-6)


fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(9,4))
ax[0].hist(scaled_costheta_1_ll, bins=50)
ax[0].set_title(f'$\cos\\theta_1$')
ax[1].hist(scaled_costheta_2_ll, bins=50)
ax[1].set_title(f'$\cos\\theta_2$')
fig.suptitle(f'$\cos\\theta$: lepton polar angle in the W rf\n Scaled data (LL data)')
```

```python
# Normalization

def normalization(arr):
    mean = np.mean(arr)*np.ones(len(arr))
    std = np.std(arr)*np.ones(len(arr))
    return (arr - mean) / std

norm_costheta_1_ll = normalization(scaled_costheta_1_ll)
norm_costheta_2_ll = normalization(scaled_costheta_2_ll)
norm_costheta_1_tl = normalization(scaled_costheta_1_tl)
norm_costheta_2_tl = normalization(scaled_costheta_2_tl)
norm_costheta_1_tt = normalization(scaled_costheta_1_tt)
norm_costheta_2_tt = normalization(scaled_costheta_2_tt)
```
